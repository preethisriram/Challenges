Histogram vs. Bar Plot?

Histograms and barplots look nearly identical, and it can be easy to mistake them in some instances. Their differences lies in what they're typically used for. Bar plots tend to measure categorical data, whereas histograms measure the frequency of numerical data.

```python
highrated = df[df['avg_rating'] >= 9.0]
```
